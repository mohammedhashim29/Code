```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_cleaned.csv')
```

```python
#summary statistics
df.describe()
```

```python
numeric_features = df.drop(columns=['Survived']).columns
```

```python
# Histograms
df[numeric_features].hist(figsize=(15, 10), bins=20)
plt.suptitle('Histograms of Numeric Features', fontsize=20)
plt.show()
```

```python

# Boxplots
plt.figure(figsize=(15,8))
sns.boxplot(data=df[numeric_features])
plt.title('Boxplots of Numeric Features')
plt.xticks(rotation=90)
plt.show()
```

```python
# Pairplot (scatter matrix)
sns.pairplot(df, hue='Survived', corner=True)
plt.suptitle('Pairplot of Features Colored by Survival', y=1.02)
plt.show()
```

```python
# Correlation Matrix
plt.figure(figsize=(12,8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Fare is positively correlated with Survived — higher paying passengers survived more.

Pclass (Passenger Class) is negatively correlated with Survived — 1st class survived more than 3rd class.

Age might show that younger passengers had a slightly higher survival chance.

Sex is highly important — females survived much more than males.

Cabin (deck) might show some relationship with survival — better decks survived more.
